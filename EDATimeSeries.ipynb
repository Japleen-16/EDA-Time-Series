```python
#Time resampling
#to find the min value for all the columns in the dataset, rule A will give us the min/max values for the whole year.

df_tesla.resample(rule='A').min()
df_tesla.resample(rule='A').max()
```

```python
# to check quaterly start results of one year we use the rule QS

df_tesla.resample(rule='QS').max()['High'].plot(figsize=(8,8))

#and hence we can infer from the graph that after 2020 we got a huge increase in the share price and the max value was between 2020-21

#for business end year
df_tesla.resample(rule='BA').max()['High'].plot(figsize=(8,8))
```

```python
#use rule=M to get the monthly data for the plot
#lets compare the monthly values year 2017 for max values of high
highOf2017=df_tesla.loc['2017-01-01':'2017-12-31']

highOf2017['High'].resample(rule='M').max().plot(figsize=(8,8),kind='bar',color='blue')
```

Simple Moving Avg

###SIMPLE MOVING AVG IS BASICALLY USED TO SMOOTHEN THE GRAPHS.

Here's a graph showing the actual open values and simple moving avg values of open using 10 days and 15 days rolling from 2021-22

```python
df_tesla.tail()
```

```python
df_tesla['Open 10 days rolling']=df_tesla['Open'].rolling(window=10,min_periods=1).mean()
df_tesla['Open 30 days rolling']=df_tesla['Open'].rolling(window=30,min_periods=1).mean()
df_tesla[['Open','Open 10 days rolling','Open 30 days rolling']].plot(xlim=['2021-01-01','2022-03-29'],figsize=(15,12),ls='--')
```

```python
#here is a plot showing the rolling for 10, 30 and 50 days for high(max) compared to the original one only for year 2020
df_tesla['High(max) rolling for 10 days']=df_tesla['High'].rolling(window=10, min_periods=1).max()
df_tesla['High(max) rolling for 30 days']=df_tesla['High'].rolling(window=30, min_periods=1).max()
df_tesla['High(max) rolling for 50 days']=df_tesla['High'].rolling(window=50, min_periods=1).max()
df_tesla[['High','High(max) rolling for 10 days','High(max) rolling for 30 days','High(max) rolling for 50 days']].plot(xlim=['2021-01-01','2021-12-31'],figsize=(15,6))
```

###Hence we infer the differences in the plots on increasing the rolling window and the use of rolling wrt to the original plot.

In SMA we give equal importance to every data but in EDA we should focus more on recent data and hence we use EWMA ie exponential weighted moving avg

```python
#CMA is done using expanding func

df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```

####EMA is used to smoothen the graphs in a better way than SMA, provides different importance to different data and works on recent data entries.
Take inital ema as sma for the given window size

EMA formula=((Close - prev EMA)*multiplier)+prevEma 
multiplier=2/(window_size+1)

EWM is the function for emwa in colab

```python
df_tesla['Open EWM_0.2']=df_tesla['Open'].ewm(alpha=0.2, adjust=False).mean() #0.3/alpha is the smoothening factor
df_tesla['Open EWM_0.5']=df_tesla['Open'].ewm(alpha=0.5,adjust=False).mean()
```

```python
df_tesla[['Open','Open EWM_0.2','Open EWM_0.5']].plot(xlim=['2020-01-01','2020-12-31'],figsize=(12,12))
#The conclusion we draw is we have to keep the alpha low in order to get a vague idea about the trends of the graph.
```
